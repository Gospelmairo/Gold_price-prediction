#  How does Adjusted Closing price evolve over the days of the month?

These lines of code generate a bar plot to visualize how the average adjusted closing price of gold evolves over the days of the month. It uses groupby() to group the data by the day of the month and then calculates the mean of the adjusted closing prices for each day

```python
# Calculate the average adjusted closing price by day of the week
average_adj_close = df.groupby('Day')['Adj Close'].mean()


# Plot the average adjusted closing price by day of the week
average_adj_close.plot(kind='bar')
plt.xlabel('Day of the month')
plt.ylabel('Average Adjusted Closing Price')
plt.title('Average Adjusted Closing Price by Day of the month')


# Display the plot
plt.show()
```

According to the graph the Adjusted closing prive evolve the most on the 26th and 27th of a month

# How does Adjusted Closing price evolve over the months of the year
This code generates a bar plot to visualize how the average adjusted closing price of gold evolves over the months of the year.

```python
df.groupby('Month')['Adj Close'].mean().plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Average Adjusted Closing Price')
plt.title('Average Adjusted Closing Price by Month')
plt.show()
```

From the graph we can see that Febuary(7) is the month where the Adjusted closing price tend to evolve the most


# What are the best days of the Month to buy and sell gold?

These line of code calculate and print the best day to buy and sell gold based on the average adjusted closing prices for each day of the month. The idxmin() and idxmax() functions are used to obtain the index (representing the day of the month) where the minimum and maximum average values occur, respectively.

```python
best_buy_day = df.groupby('Day').mean().idxmin()
best_sell_day = df.groupby('Day').mean().idxmax()       
print("Best day to buy gold:", best_buy_day)
print("Best day to sell gold:", best_sell_day)
```

```python
df[ ["Adj Close", "Day"] ].groupby("Day").mean().plot(kind = 'bar')
plt.xlabel('Day')
plt.ylabel('Average Adjusted Closing Price')
plt.title('Best Day Of The Month To Buy And Sell Gold')
plt.show()
```

The best day of the month to buy gold is when the prices is at the lowest which is 29th of a month
The best day of the month to sell gold is when the prices is at the highest which is the 27th of a month

# What are the best months of the year to buy and sell gold?

These lines calculate and print the best month to buy and sell gold based on the average adjusted closing prices for each month. The idxmin() and idxmax() functions are used to obtain the index (representing the month) where the minimum and maximum average values

```python
adjusted_closing_price = df.groupby('Month')['Adj Close'].mean()

best_buy_month = adjusted_closing_price.idxmin()
best_sell_month = adjusted_closing_price.idxmax()
print("Best month to buy gold:", "Month", best_buy_month)
print("Best month to sell gold:", "Month", best_sell_month)
```

# How does the difference between Close and Adjusted Close evolve over the days of the month?

```python
# calculates the difference between the closing price and the adjusted closing price for each row in the DataFrame
# And store it in a new column called 'close_adjusted_diff'.


df['close_adjusted_diff'] = df['Close'] - df['Adj Close']
df['close_adjusted_diff'].head()
```

```python
# Descriptive statistics of the 'close_adjusted_diff' column using the describe() function and print the results

df['close_adjusted_diff'].describe()
```

```python
# Plot the average difference over the days of the month and make the plot on a log scale

average_diff_by_day = df.groupby('Day')['close_adjusted_diff'].mean()

if average_diff_by_day.min() > 0:
    plt.yscale('log')

plt.plot(average_diff_by_day)
plt.xlabel('Day of the week')
plt.ylabel('Average Difference (Close - Adjusted Close)')
plt.title('Average Difference (Close - Adjusted Close) by Day of the month')

# Check if the minimum value of the average differences is greater than zero again. If it is,
# set the x-axis and y-axis limits for better visualization.

if average_diff_by_day.min() > 0:
    plt.xlim(0, 10)
    plt.ylim(-100, 100)

plt.show()
```

from the graph we can see that theres no difference between the 'Close' and 'Adj Close' in the monthly evolvement

### Analyze the correlation between variables:
This will create a heatmap to visualize the correlation between different variables in the dataset.

With specific columns

1. 'Open': The opening price of the day might provide insights into the initial market sentiment.
2. 'High': The highest price of the day could indicate significant market activity or positive sentiment.
3. 'Low': The lowest price of the day might reflect market volatility or negative sentiment.
4. 'Close': The closing price represents the final trading value and can provide overall market sentiment.
5. 'Volume': The trading volume could indicate market participation and liquidity.
6. 'Day'


```python
# Select the desired columns
selected_columns = ['Day', 'Open', 'High', 'Low', 'Close', 'Volume']

# Create the correlation matrix
corr_matrix = df[selected_columns].corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10, 10))  # Adjust the figure size as desired
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
